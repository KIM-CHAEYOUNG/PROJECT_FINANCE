# **기사제목**
### 텍스트 전처리

```python
df.Title
```

```python
import re
pattern=r'[^가-힣]'
ntitles=[]
for t in df.Title:
    rtitle=re.sub(pattern,'', t)
    
    ntitles.append(rtitle)
print(ntitles[:10])
```

# 토큰화

```python
import konlpy
from konlpy.tag import Kkma
kkma=Kkma()
```

```python
ntitles
```

```python
kpos=kkma.pos(ntitles[0])
kpos
```

```python
worddic={}
for t in ntitles:
    kpos=kkma.pos(t)
    jlist=[]
    jlist.append(kpos)
    for pos in kpos:
        j='|'.join(pos)
        if(j in worddic.keys()):
            worddic[j]+=1
        else:
            worddic[j]=1
```

```python
print(worddic)
```

```python
kkmaTag=[]
for i in ntitles:
    kkmaTag+=kkma.pos(i)
print(kkmaTag[:20])
```

```python
from collections import Counter
Counter(kkmaTag).most_common()
```
