```python
urls=[]
TITLE=[]
CONTENT=[]
DATE=[]
#수빈님 코드 참고
URL="http://stock.thinkpool.com/bbs/itemanal/list/stock_bbs.do?pageNo={}&pUd=0&iMax=00110606909999&code=035720"
for i in range(1,3):
    NURL=URL.format(i)
    res=requests.get(NURL)
    soup=BeautifulSoup(res.content,'html.parser')
    tgts=soup.select('.bbs_list tbody tr a')
    for tgt in tgts:
        url=tgt.attrs['href']
        if(url[:4]=='/bbs'):
            urls.append('http://stock.thinkpool.com/'+url)
            for url in urls:
                #글 하나의 url 파싱
                res2=requests.get(url)
                soup2=BeautifulSoup(res2.content,'html.parser')
                #제목
                title=soup2.find('div', class_='head_area').find('strong').get_text()
                TITLE.append(title)
                #내용
                doc=soup2.select('div.tx_area')[0].text.strip()
                endpos=doc.find("\n\n\n\n\n\n")
                doc2=doc[:endpos]
                CONTENT.append(doc2)
                #날짜
                date=soup2.select('.info_area')[0].text
                ndate=date[-25:-12].strip().replace('/','')
                DATE.append(ndate)

                print(ndate)
                print(title)
                print(doc2)
                print('*'*30)
```
